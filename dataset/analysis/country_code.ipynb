```python
import pandas as pd
import matplotlib.pyplot as plt

# CSVファイルを読み込む

# ".com"で終わる行だけをフィルタリング
# df = df[df['domain'].str.endswith('.com')]
tlds = ["", 'nl', 'com', 'net', 'de', 'org', 'eu']

for tld in tlds:
    df = pd.read_csv("hall-of-flame-websites-tlsa-usage3-ip-address-country-code.csv")
    if tld != "":
        df = df[df['domain'].str.endswith('.' + tld)]
    # country_codeごとの割合を計算
    country_counts = df['country_code'].value_counts().head(10)
    total_count = len(df)
    country_percentages = country_counts / total_count * 100

    # グラフを作成
    plt.figure(figsize=(10, 6))
    country_percentages.plot(kind='bar', color='skyblue')
    # if tld == "":
    #     plt.title('Country Code Percentage')
    # else:
    #     plt.title(f'Country Code Percentage ({tld})', fontsize='x-large')
    plt.xlabel('Country Code', fontsize='x-large')
    plt.ylabel('Percentage', fontsize='x-large')
    plt.xticks(rotation=0)  # x軸のラベルを回転させて読みやすくする
    plt.gca().xaxis.set_ticks_position('none')

    plt.tight_layout()

    # グラフを表示
    plt.savefig(f'country_code_percentage_{tld}.pdf')
```
